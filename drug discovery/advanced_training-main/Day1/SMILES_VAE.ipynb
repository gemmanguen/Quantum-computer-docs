```
import torch
import torch.nn as nn
import torch.nn.functional as F
class AttrPredictor(nn.Module):
  """
  Receives latent vector z, predicts LogP, SA, and QED.
  """
  
  def __init__(self, d_z):
    super().__init__()
    self.predictors = nn.ModuleDict(
        {k: nn.Sequential(
            nn.Linear(d_z, d_z),
            nn.ReLU(),
            nn.Linear(d_z, 1)) 
        for k in ['logp', 'sa', 'qed']})
  
  def forward(self, z):
    return [self.predictors[k](z) for k in ['logp', 'sa', 'qed']]
```

```
%%time

def pmap(pickleable_fn, data, n_jobs=cpu_count() - 1, verbose=1, **kwargs):
    """
    Parallel map using joblib.
    :param pickleable_fn: Fn to map over data.
    :param data: Data to be mapped over.
    :param kwargs: Additional args for f
    :return: Mapped output.
    """
    return Parallel(n_jobs=n_jobs, verbose=verbose)(
        delayed(pickleable_fn)(d, **kwargs) for d in data
    )

# %%

mhfp_enc = MHFPEncoder()
lshforest = LSHForestHelper()


def add_to_forest(idxed_smiles):
    idx, smiles = idxed_smiles
    fp = mhfp_enc.encode(smiles)
    return fp


fps = pmap(add_to_forest, enumerate(moses_seed[:100000]))
# ~250k smiles / min

for i, fp in tqdm(enumerate(fps)):
    lshforest.add(i, fp)

lshforest.index()

pain_fps = [mhfp_enc.encode(s) for s in pain_set["smiles"]]

nns = set()
for fp in pain_fps:
    nns.update(set(lshforest.query(fp, k=200, data=fps)))
    print(len(nns))

moses_seed.iloc[list(nns)].to_csv("knn_200_from_moses_seed.csv")
# 63899

# %%
#moses_train = pd.read_csv("~/data/_moses_train.csv")
#zs = moses_seed.iloc[list(nns)].rename(columns={"smiles": "SMILES"})

#df = moses_train.append(zs)
#df["SPLIT"] = "train"
#df.to_csv("~/data/knn_1000_from_zinc_train.csv", index=False)
```

```
def cal_prop(smiles_list):
  success_smiles = []
  logP_list = []
  qed = []
  sas = []
  count = 0
  z_num = []
  for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
      try:
        logp = logP(mol)
        qd = QED(mol)
        sa = SA(mol)
        success_smiles.append(smiles)
        logP_list.append(logp)
        qed.append(qd)
        sas.append(sa)
        z_num.append(count)
        #print(smiles, logp, qd, sa)
      except:
        pass
    count += 1
  return success_smiles, logP_list, qed,sas, z_num

def get_prop(smiles_list) :
  smi,l,q,s,z_num = cal_prop(smiles_list)
  df = [smi,l,q,s]
  df = pd.DataFrame(df).T
  df.columns = ['smiles','logP','QED','SAS']
  return df, z_num
```

```
%%time
new_sample_prop, new_z_num= get_prop(new_sample)
print(new_sample_prop)
```

```
%%time

old_sample_prop, old_z_num= get_prop(old_sample)
print(old_sample_prop)
```
